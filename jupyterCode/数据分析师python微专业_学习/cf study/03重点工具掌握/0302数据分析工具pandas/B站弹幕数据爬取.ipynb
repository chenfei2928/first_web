```python
import requests

url ='https://comment.bilibili.com/10959711.xml'

r= requests.get(url)

r
```

```python
r.encoding='utf-8'   # 有中文用此编码方式
print(r.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text,'lxml')
soup
```

```python
ds = soup.find_all('d')
print(len(ds))
```

```python
ds[0]
```

```python
ds[0].text
```

```python
ds[0]['p'].split(',')[4]
```

```python
datalst =[]
for d in ds:
    dic = {}
    dic['弹幕'] = d.text
    dic['时间'] = datetime.fromtimestamp(int( d['p'].split(',')[4]))
    datalst.append(dic)
datalst
```

```python
import pandas as pd
df = pd.DataFrame(datalst)
df
```
