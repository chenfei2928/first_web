```python
"""
Pandas时间戳索引：DatetimeIndex

核心：pd.date_range()

"""
```

```python
import pandas as pd
import numpy as np
```

```python
# pd.DatetimeIndex() 与 TimeSeries时间序列

rng = pd.DatetimeIndex(['12/1/2017','12/2/2017','12/3/2017','12/4/2017','12/5/2017'])
print(rng,type(rng))

print(rng[0],type(rng[0]))
    # 直接生成时间戳索引，支持str、datetime
    # 单个时间戳为Timestamp，多个时间戳为DatetimeIndex

st = pd.Series(np.random.rand(len(rng)),index=rng)
print(st,type(st))
print(st.index)
    # 以DatetimeIndex为index的Series，为TimeSeries，时间序列
```

```python
# pd.date_range()  日期范围，生成日期范围
# 2种生成方式 ① start+end ; ② start/end + periods
# 默认频率：day

rng1 = pd.date_range('1/1/2017','1/10/2017',normalize=True)
print(rng1,type(rng1))
print('...........................分割线..1..............................................................')
rng2 = pd.date_range(start='1/1/2017',periods=10)
print(rng2)
print('...........................分割线..2..............................................................')
rng3 = pd.date_range(start='1/30/2017 15:00:00',periods=10)
print(rng3)
# 直接生成DatetimeIndex
# pd.date_range
    #(start=None, end=None, periods= None,freq='Q',ts=None,normalize=False,name=None, closed=None,** kwargs为了兼容性。对结果没有影响。 )
# start 开始时间
# end 结束时间
# periods 偏移量
# freq 频率，默认天，pd.date_range()默认频率为日历日，pd_bdate_range()默认频率为工作日
# ts 时区

print('...........................分割线..3..............................................................')

rng4 = pd.date_range(start = '1/1/2017 15:30',periods = 10 ,name = 'hello world',normalize =True)
print(rng4)

# normalize ： bool，默认为False  在生成日期范围之前将开始/结束日期标准化为午夜。
# name：索引对象名称
print('...........................分割线..4..............................................................')

print(pd.date_range('20170101','20170104'))
print(pd.date_range('20170101','20170104',closed = 'right'))
print(pd.date_range('20170101','20170104',closed = 'left'))
# closed: 默认为None的情况下，左闭右闭，left则左闭右开 ，right则左开右闭

print('...........................分割线..5..............................................................')

print(pd.bdate_range('20190101','20190107'))
   # pd.bdate_range() 默认频率为工作日

print(list(pd.date_range(start='1/1/2017',periods=10)))
# 直接转化为list，元素为Timestamp
```

```python
# pd.date_range() 日期范围 频率
print(pd.date_range('2019/1/1','2019/1/5'))   # 默认freq = ‘D’ 每日历日      
print(pd.date_range('2019/1/1','2019/1/5',freq='B'))   #B 每工作日
print(pd.date_range('2019/1/1','2019/1/2',freq='H'))   #B 每小时
print(pd.date_range('2019/1/1 12:00','2019/1/1 12:10',freq='MIN')) # T/MIN 每分；
print(pd.date_range('2019/1/1 12:00:00','2019/1/1 12:00:09',freq='S')) # S 每秒
print(pd.date_range('2019/1/1 12:00:00','2019/1/1 12:00:09',freq='L')) # L  每毫秒（千分之一秒）
print(pd.date_range('2019/1/1 12:00:00','2019/1/1 12:00:09',freq='U')) # U  每微秒（百万分之一秒）

print(pd.date_range('2019/1/1','2019/2/1',freq='W-MON'))
    # W-MON 从指定星期几开始算起，每周
    # 星期几缩写 :MON/TUE/WED/THU/FRI/SAT/SUN

print(pd.date_range('2019/1/1','2019/4/1',freq='WOM-2MON'))
    # WOM-2MON; 每月的第几个星期几开始算，这里是每月第二个星期一

```

```python
# pd.date_range()  日期范围，频率（2）
print(pd.date_range('2017','2018',freq='M'))  # 每月最后一个日历日
print(pd.date_range('2017','2020',freq='Q-DEC')) # Q-月 ；指定月为季度末，每个季度末最后一月的最后一个日历日
print(pd.date_range('2017','2020',freq='A-DEC'))
    # A-月：每年指定月份的最后一个工作日
    # 月缩写 JAN\FEB\MAR\APR\MAY\JUN\JUL\AUG\SEP\OCT\NOV\DEC
print('-------------------------------分割线1--------------------------------------------------')
   
print(pd.date_range('2017','2018',freq ='BM'))  #每月最后一个工作日
print(pd.date_range('2017','2020',freq='BQ-DEC')) # BQ-月 ；指定月为季度末，每个季度末最后一月的最后一个工作日
print(pd.date_range('2017','2020',freq='BA-DEC')) # BA-月，每年指定月份的最后一个工作日
print('-------------------------------分割线2--------------------------------------------------')

print(pd.date_range('2017','2018',freq='MS'))  # 每月第一个日历日
print(pd.date_range('2017','2020',freq='QS-DEC')) # QS-月 ；指定月为季度末，每个季度末最后一月的第一个日历日
print(pd.date_range('2017','2020',freq='AS-DEC'))  #AS-月： 每年指定月份的第一个工作日

print('-------------------------------分割线3--------------------------------------------------')
print(pd.date_range('2017','2018',freq ='BMS'))  #每月第一个工作日
print(pd.date_range('2017','2020',freq='BQS-DEC')) # BQS-月 ；指定月为季度末，每个季度末最后一月的第一个工作日
print(pd.date_range('2017','2020',freq='BAS-DEC')) # BAS-月，每年指定月份的第一个工作日
```

```python
# pd.date_range()   日期范围，复合频率

print(pd.date_range('2019/1/1','2019/2/1',freq='7D'))  #7天；
print(pd.date_range('2019/1/1','2019/1/2',freq='2h30min'))  #2小时30分钟
print(pd.date_range('2017','2018',freq='2M'))# 2月，每月最后一个日历日
```

```python
# asfreq 时期频率转换

ts = pd.Series(np.random.rand(4),
               index=pd.date_range('20170101','20170104'))
print(ts)

print(ts.asfreq('4H',method='ffill'))
    # 改变频率，这里D改为4H
    # method: 插值模式，None不插值，ffill用之前值填充，bfill用之后值填充
```

```python
# pd.date_range()  日期范围  超前/滞后数据

ts = pd.Series(np.random.rand(4),
              index=pd.date_range('20190101','20190104'))
print(ts)
print(ts.shift(2))
print(ts.shift(-2))
    # 正数：数值后移（滞后）；负数：数值前移（超前）
print('......................................................')
per = ts/ts.shift(1)-1
print(per)
    # 计算变化百分比，这里计算，该时间戳与上一个时间戳相比，变化百分比；
print('......................................................')

print(ts.shift(2,freq='D'))
print(ts.shift(2,freq='T'))
    # 加上Freq参数；对时间戳进行位移，而不是对数值进行位移
```
