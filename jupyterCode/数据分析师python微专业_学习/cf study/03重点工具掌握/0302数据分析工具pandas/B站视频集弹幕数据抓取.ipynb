```python
cids_raw = text.split('"cid":')
cids_raw
```

```python

cids = [x[:8] for x in cids_raw if len(x)>8]
cids
```

```python
cids=list(set(cids))
len(cids)
```

```python
def cid_to_ul(cid):
    return 'https://comment.bilibili.com/%s.xml'%cid

cid_to_ul('10959724')
```

```python
def url_to_data(url):
    r = requests.get(url)
    r.encoding = 'utf-8'  # 防止乱码
    
    soup = BeautifulSoup(r.text,'lxml')
    ds = soup.find_all('d')
    
    datalist=[]
    for d in ds:
        dic={}
        dic['弹幕']=d.text
        dic['时间']=datetime.fromtimestamp(int( d['p'].split(',')[4]))
        datalist.append(dic)
    return datalist

url_to_data('https://comment.bilibili.com/10959724.xml')
```

```python
def data_inte(cids):
    urls = [cid_to_ul(cid) for cid in cids]
    alldata =[]
    for url in urls:
        datalist = url_to_data(url)
        alldata.extend(datalist)
        print('已添加%s条弹幕数据！'% len(alldata))
    return alldata

alldata=data_inte(cids)
```

```python
df = pd.DataFrame(alldata)
df
```

```python
df.to_csv('弹幕数据汇总3.csv',index=False,encoding='utf_8_sig')
```
