```python
'''
DataFrame：基本技巧
数据查看、转置、添加、修改、删除值、对齐、排序

'''
```

```python
import numpy as np
import pandas as pd
```

```python
# 数据查看、转置

df = pd.DataFrame(np.random.rand(16).reshape(8,2)*100,
                  columns=['a','b'])
print(df)

print(df.head(2))

print(df.tail())
    # .head()查看头部数据
    # .tail()查看尾部数据
    # 默认查看5条
print(df.T)
    # .T 转置
```

```python
# 添加与修改

df = pd.DataFrame(np.random.rand(16).reshape(4,4)*100,
                   columns = ['a','b','c','d'])
print(df)

df['e']=10
df.loc[4]=22
print(df)
    # 新增行列并赋值

df['e']=20
df[['a','c']]=100
print(df)
    # 索引后直接修改值
```

```python
# 删除  del / drop()

df = pd.DataFrame(np.random.rand(16).reshape(4,4)*100,
                   columns = ['a','b','c','d'])
print(df)

del df['a']
print(df)
    #del语句  删除列
print('..........')
print(df.drop(0))
print(df.drop([1,2]))
print(df)
    # drop()删除行，inplace=False   删除后生成新的数据，不改变原数据
print('...........................')
print(df.drop(['d'],axis=1))
print(df)
    # drop()删除列，需要加上axis=1,inplace=False   删除后生成新的数据，不改变原数据
```

```python
# 对齐

df1 = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
df2 = pd.DataFrame(np.random.rand(7,3),columns =['a','b','c'])

print(df1)
print(df2)
print(df1+df2)
    # DataFrame 对象之间的数据自动按照列和索引（行标签）对齐
```

```python
# 排序1 
#      按值排序.sort_values
#      同样适用于Series
df1 = pd.DataFrame(np.random.rand(16).reshape(4,4)*100,
                  columns=['a','b','c','d'])
print(df1)

print(df1.sort_values(['a'],ascending=True))    # a列升序排列
print(df1.sort_values(['a'],ascending=False))    # a列降序排列
    # ascending参数，设置升序降序，默认升序
    #单列排序
print('.......fengexian...............................................')

df2 = pd.DataFrame({'a':[1,1,1,1,2,2,2,2],
                  'b':list(range(8)),
                  'c':list(range(8,0,-1))})
print(df2)
print('.......................................')
print(df2.sort_values(['a','c']))
    # 多列排序，按列顺序排序
```

```python
# 排序2 
#      索引排序 .sort_index
df1 = pd.DataFrame(np.random.rand(16).reshape(4,4)*100,
                  index = [5,4,3,2],
                   columns = ['a','b','c','d'])
print(df1)
print(df1.sort_index())
    #按照index排序
df2 = pd.DataFrame(np.random.rand(16).reshape(4,4)*100,
                  index = ['h','s','x','g'],
                   columns = ['a','b','c','d'])
print(df2)
print(df2.sort_index())
    #按照index排序
    #默认ascending =True 升序，inplace =False
```

```python
#作业1：创建一个3*3，值在0-100区间随机值的Dataframe（如图），分别按照index和第二列值大小，降序排序

df =pd.DataFrame(np.random.rand(9).reshape(3,3)*100,
                index=['b','a','c'],
                columns=['v1','v2','v3'])
print(df)

print(df.sort_index(ascending=False))
print(df.sort_values(['v2'],ascending=False))
```

```python
#作业2：创建一个5*2，值在0-100区间随机值的Dataframe（如图）df1，通过修改得到df2

df = pd.DataFrame(np.random.rand(10).reshape(5,2)*100,
                 index=['a','b','c','d','e'],
                 columns=['v1','v2'])
print(df)

df2 = df.drop(['e'])
print(df2)

df3= df2.T
print(df3)
```
