```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(np.random.randn(50,4),
                 index = pd.date_range('2019-1-1',periods=50),
                 columns=list('ABCD'))
    # numpy.random.randn(d0, d1, …, dn)是从标准正态分布中返回一个或多个样本值。 
    # numpy.random.rand(d0, d1, …, dn)的随机样本位于[0, 1)之间。 

df
```

```python
df = df.cumsum()
df
```

```python
# 折线图

df.plot(kind='line',
       style='--o',
       figsize=(12,6),
       grid=True,
       colormap='copper')
```

```python
# 散点图，气泡图

plt.figure(figsize=(12,8))
plt.scatter(df['A'],df['B'],
           marker='x',
           s = df['C'].abs()*100,
           alpha=0.8,
           c= df['D'],
           cmap ='Reds')
```

```python
# 箱型图

df.boxplot(figsize=(12,8),grid=False)
```

```python
import seaborn as sns

sns.set_context('talk')
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(12,8))

sns.violinplot(x='day',y='total_bill',data=tips,
                order=['Sun','Sat','Fri','Thur'],
              scale = 'count')
```

```python
plt.figure(figsize=(12,8))

sns.kdeplot(df['A'],df['B'],
           shade=True,
           cmap='Reds',
           shade_lowest = False)
```

```python
plt.figure(figsize=(12,8))

sns.jointplot(df['A'],df['B'],
             marginal_kws={'bins':15})
```
