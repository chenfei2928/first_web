1、导入需要使用到的库

```python
import jieba
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from imageio import imread
```

2、读取excel文件内容

```python
def get_wordList():
    df = pd.read_excel('C:\\Users\\Administrator\\Desktop\\python\\完美陌生人-短评.xlsx')
    wordList = df['评论内容'].tolist()
    return wordList
```

```python
def get_wordClound(mylist):
    word_list = [" ".join(jieba.cut(sentence)) for sentence in mylist]
    new_text = ' '.join(word_list)
    pic_path = 'C:\\Users\\Administrator\\Desktop\\python\\guniang.jpg'
    img_mask = imread(pic_path)

    wordcloud = WordCloud(background_color="white", 
                          font_path='msyh.ttc',
                          mask=img_mask, stopwords=STOPWORDS, ).generate(new_text)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
```

```python
if __name__ == '__main__':
    wordList = get_wordList()
    get_wordClound(wordList)
```
