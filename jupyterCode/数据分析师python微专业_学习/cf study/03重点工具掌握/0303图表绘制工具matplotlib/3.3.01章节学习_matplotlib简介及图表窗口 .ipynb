```python
'''
Matplotlib  一个python的matlab绘图接口，支持 python、numpy 、 pandas 基本数据结构。运营高效且具有较丰富的图表库
'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 图表窗口    plt.show()

plt.plot(np.random.rand(10))
plt.show()
```

```python
  #   图表窗口02 — 魔法函数  嵌入图表
    
%matplotlib inline

x = np.random.rand(1000)
y = np.random.rand(1000)

plt.scatter(x,y)
    # 直接嵌入图表，不用plt.show()
    
```

```python
#图表窗口3 —魔法函数，弹出可交互的matplotlib窗口

%matplotlib notebook
s = pd.Series(np.random.randn(100))
s.plot(style='k--o',figsize=(10,5))
    # 可交换的matplotlib窗口，不用plt.show()
    # 可做一定调整
```

```python
 # 图表窗口4 —魔法函数，弹出matplotlib控制台

%matplotlib qt5
df = pd.DataFrame(np.random.rand(50,2),columns=['A','B'])
df.hist(figsize=(12,5),color='g',alpha=0.8)

# 可交换性控制台
# 如果已经设置了显示方式，（比如notebook），需要重启然后再运行魔法函数
# 网页嵌入的交互性窗口和控制台，只能显示一个

# plt.close

plt.gcf().clear()
 # 每次清空图表内容

# 自己练习时，若不出现交换性窗口，可将此段代码单独复制一窗口运行
```
