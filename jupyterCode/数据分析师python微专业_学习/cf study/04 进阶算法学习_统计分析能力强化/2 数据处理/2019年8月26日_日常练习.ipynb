isin函数，这个函数能够帮助我们快速判断源数据中某一列（Series）的值是否等于列表中的值。拿案例来说，df['流量来源'].isin(['二级','三级'])，判断的是流量来源这一列的值，是否等于“二级”或者“三级”，如果等于（等于任意一个）就返回True，否则返回False。我们再把这个布尔型判断结果传入行参数，就能够很容易的得到流量来源等于二级或者三级的渠道。

```python
#场景三：我们想要提取二级、三级流量来源、来源明细对应的访客和支付转化率。

data.loc[data['流量来源'].isin(['二级','三级']),['流量来源','来源明细','访客数','支付转化率']]
```

```python
data['访客数'].mean()  #平均值
data['访客数'].std()   # 标准差
data['访客数'].median()  # 中位数
data['访客数'].max()    # 最大值
data['访客数'].min()    # 最小值
```

```python
#场景四：对于流量渠道数据，我们真正应该关注的是优质渠道，假如这里我们定义访客数、转化率、客单价都高于平均值渠道是优质渠道，
#那怎么找到这些渠道呢？

print('访客数均值',data['访客数'].mean())
print('转化率均值',data['支付转化率'].mean())
print('客单价均值',data['客单价'].mean())
```

```python
data['访客数'] > data['访客数'].mean()
```

```python
data.loc[(data['访客数'] > data['访客数'].mean()) & 
          (data['支付转化率'] > data['支付转化率'].mean()) & 
          (data['客单价'] > data['客单价'].mean()),:]
```
