```python
import requests
from bs4 import BeautifulSoup
```

```python
urllist=[]
for i in range(0,226,25):
    list=('https://movie.douban.com/top250?start=%i&filter=' %i)
    urllist.append(list)
print(urllist)
    # 每页25个，10页提取成功
```

```python
data =[]
for url in urllist:
    r = requests.get(url)     #页面访问
    soup =BeautifulSoup(r.text,'lxml') #解析网页
    div = soup.find('div',class_="article")
    lis = div.find_all('li')
    for li in lis:
        dic ={}
        dic['电影名称'] = li.find('span',class_="title").text
        dic['豆瓣评分'] = li.find('span',class_="rating_num",property="v:average").text
        dic['评价人数'] = li.find('div',class_="star").find_all('span')[-1].text[0:-3]
        dic['电影短评'] = li.find('span', class_="inq")
        data.append(dic)
        
data
```
