存储目标信息

```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.runoob.com/html/html-intro.html'
r = requests.get(url)

#html = r.text.encode(r.encoding).decode()
    # 编码方式，此行代码不太理解
html = r.content.decode("utf-8")
        
soup = BeautifulSoup(html,'lxml')
soup.head()
```

```python
#l = soup.find_all('h2')
#l = soup.findAll('h2')   与第一行代码一样效果
import pandas as pd

l = [x.text for x in soup.find_all('h2')]   # 列表生成式
#l
l
```

```python
df =pd.DataFrame(l,columns=[url])
df.to_excel('爬虫.xlsx')
```

```python
df.iloc[2,0]
```

```python
df.to_csv('爬虫.csv',encoding='utf_8_sig')
    # 生成csv文件乱码 
```

##爬取其他网页

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
x = soup.find_all('a')[1]
x.has_attr('href')
x.attrs['href']
```

```python
links =[i for i in soup.find_all('a') 
        if i.has_attr('href') and i.attrs['href'][0:5]=='/html']

relative_urls = set([i.attrs['href'] for i in links])

absolute_urls = {'http://www.runoob.com' + i for i in relative_urls}

absolute_urls.discard(url)
absolute_urls
```

```python
for i in absolute_urls:
    ri = requests.get(i)    
    soupi = BeautifulSoup(ri.content.decode("utf-8"),'lxml')
    li = [x.text for x in soup.find_all('h2')]   # 列表生成式
    dfi = pd.DataFrame(li,columns=[i])
#    print(dfi)
    df = df.join(dfi, how='outer')

df
```

```python
df.to_excel('爬虫多网页.xlsx')
```
