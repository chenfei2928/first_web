# 数据清洗

## 1、双表合并

```python
moive_combine = pd.merge(base_info,moive_info,left_on = '网址',right_on='网址',how='inner')
```

```python
moive_combine.head()
    # 合并后数据似乎重复了？需要搞清楚
```

## 2、上映时间清洗

```python
moive_combine['上映时间'].head()
```

```python
a = moive_combine['上映时间'].str.split('').str[2:6]
```

```python
moive_combine['上映时间'] = a.str[0]+a.str[1]+a.str[2]+a.str[3]
```

```python
moive_combine['上映时间'].head()
```

## 3、评分分布调整

```python
# 查看评分分布字段的格式
print(type(moive_combine['评分分布'].values[0]))
moive_combine['评分分布'].values[0]
```

```python
# 单引号替换为双引号

var = moive_combine['评分分布'].values[0].replace("\'",'\"')
var
```

```python
type(json.loads(var))
```

```python
# 批量对评分进行操作

def transform(x):
    return json.loads(x.replace("\'",'\"'))

moive_combine['format_评分'] = moive_combine['评分分布'].apply(transform)
```

```python
# 把单列字典的评分分布转化成分开的5列，且每一列都是数值型的

def get_rate(x,types):
    try:
        return float(x[types].strip('%'))
    except:
        pass

moive_combine['5星'] = moive_combine['format_评分'].apply(get_rate,types='力荐')
moive_combine['4星'] = moive_combine['format_评分'].apply(get_rate,types='推荐')
moive_combine['3星'] = moive_combine['format_评分'].apply(get_rate,types='还行')
moive_combine['2星'] = moive_combine['format_评分'].apply(get_rate,types='较差')
moive_combine['1星'] = moive_combine['format_评分'].apply(get_rate,types='很差')
```

```python
moive_combine.head()
```

# 数据分析



```python
# 按照总评分，5星评分人数占比，4星占比，3星.. 依次类推

moive_combine.sort_values(['5星','4星','3星','2星','1星'],ascending = False,inplace = True)
moive_combine.head()
```

## 1、剔除掉没有评分的电影并绘制

```python
moive_combine = moive_combine.loc[moive_combine['评分人数']!='None',:]
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,5))

plt.rcParams['font.sans-serif'] = ['SimHei']
# Matplotlib中设置字体-黑体，解决Matplotlib中文乱码问题
plt.rcParams['axes.unicode_minus'] = False
# 解决Matplotlib坐标轴负号'-'显示为方块的问题

sns.distplot(moive_combine['评分人数'].values,color = 'red',kde = False,bins = 200,ax = ax)
#sns.distplot(moive_combine['评分人数'].values,color = 'red',kde = False,bins = 200,ax = ax)

ax.axvline(moive_combine['评分人数'].mean(),color = 'red',alpha =0.7)
ax.axvline(moive_combine['评分人数'].median(),color='blue',alpha=0.5)
plt.xlim(0,600000)
plt.yticks(fontsize=11)
plt.xticks(fontsize=11)
ax.set_xlabel('评分人数',fontsize=14)
ax.set_ylabel('数量',fontsize=13)
```

## 2、用中位数作为人数筛选机制

```python
print('中位数:',moive_combine['评分人数'].median())
print('均值',moive_combine['评分人数'].mean())
moive_f = moive_combine.loc[moive_combine['评分人数'] > moive_combine['评分人数'].median(),: ]
```
