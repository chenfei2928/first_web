```python
'''
【项目01】  商铺数据加载及存储

作业要求：
1、成功读取“商铺数据.csv”文件
2、解析数据，存成列表字典格式：[{'var1':value1,'var2':value2,'var3':values,...},...,{}]
3、数据清洗：
① comment，price两个字段清洗成数字
② 清除字段缺失的数据
③ commentlist拆分成三个字段，并且清洗成数字
4、结果存为.pkl文件

    '''
```

```python
path = 'C:\\Users\Administrator\\Desktop\\python\\商铺数据.csv'
f = open(path,'r',encoding ='utf-8')
for i in f.readlines()[:10]:
    print((i.split(',')[-1]).split('                                '))
```

```python
# 清洗数据
    
def fcom(s):
    if '条'in s:
        return(int(s.split(' ')[0]))
    else:
        return('缺失数据')
def fpri(p):
    if '￥' in p:
        return(int(p.split('￥')[-1]))
    else:
        return('缺失数据')

def fcl(cl):
    if  len(cl)==3:
        quality = float(cl[0][2:])
        envirment =float(cl[1][2:])
        service = float(cl[2][2:])
        return([quality,envirment,service])
    else:
        return('缺失数据')
f.seek(0)      
for i in f.readlines()[:10]:
    #cm = fcom(i.split(',')[2])
    #print(cm)
    
    ww = fcl(i.split(',')[-1].split('                                '))
    print(ww)
```

```python
 # 数据清洗
datalist = []
f.seek(0)
n=0
for i in f.readlines()[:]:
    data=i.split(',')
    classify =data[0]
    name=data[1]
    com_count=fcom(data[2])
    star = data[3]
    price= fpri(data[4])
    addr = data[5]
    qua = fcl(data[-1].split('                                '))[0]
    env = fcl(data[-1].split('                                '))[1]
    service =fcl(data[-1].split('                                '))[2]
   # print(classify,name,com_count,star,price,addr,qua,env,service)
    if '缺失数据' not in [com_count,price,qua]:
        n+=1
        datalist2=[['classify',classify],
                  ['name',name],
                  ['com_count',com_count],
                  ['star',star],
                  ['price',price],
                  ['addr',addr],
                  ['qua',qua],
                  ['env',env],
                  ['service',service]]
        #print(datalist2)
        datalist.append(dict(datalist2))
        print('成功加载%i条数据'%n)
        print(dict(datalist2))
    else:
            continue
f.close()
print(datalist)
print('共加载%i条数据!'%n)
            
            
```

```python
import pickle

pic = open('C:\\Users\\Administrator\\Desktop\\python\\dat.pkl','wb')
pickle.dump(datalist,pic)
pic.close()


p = open('C:\\Users\\Administrator\\Desktop\\python\\dat.pkl','rb')

data2 = pickle.load(p)
print(data2)
```
