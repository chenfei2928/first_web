```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['SimHei']  # 解决matplotlib中 中文不显示问题；
```

```python
a =list(np.random.randint(78,150,250))
# 100到150分钟
```

```python
d =5  #组距

numbins = max(a)-min(a)//d
```

```python

fig = plt.figure(figsize=(20,8),dpi=80)

#plt.hist(a,numbins)# 频数分布直方图
plt.hist(a,numbins,normed=True)# 频率分布直方图


        # 分成20组
    
plt.xticks(range(min(a),max(a)+d,d))

plt.grid()
plt.show()
```
