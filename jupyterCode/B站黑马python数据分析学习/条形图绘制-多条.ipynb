```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['SimHei']  # 解决matplotlib中 中文不显示问题；
```

```python
a = ["猩球崛起3:终极之战",'敦刻尔克','蜘蛛侠:英雄归来','战狼2']

b16 = [15746,312,4497,319]
b15 = [12357,156,2045,168]
b14 = [2358,399,2358,362]


x14 =list(range(len(a)))
bar_width = 0.2
x15 =[i+bar_width for i in  x14]
x16 =[i+bar_width*2 for i in  x14]

fig = plt.figure(figsize=(20,8),dpi=80)

plt.bar(range(len(a)),b14,width=0.2,label='9月14日')
plt.bar(x15,b15,width=0.2,label='9月15日')
plt.bar(x16,b16,width=0.2,label='9月16日')

plt.xticks(x15,a)
plt.legend()
plt.show()
```
