3月份，10月份温度统计；


```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['SimHei']  # 解决matplotlib中 中文不显示问题；
```

```python
y3 = sorted(list(np.random.randint(1,25,31)))

y10 = sorted(list(np.random.randint(1,35,31)),reverse=True)

x3= range(1,32)
x10 = range(51,82)
fig = plt.figure(figsize = (20,8),dpi=80)

plt.scatter(x3,y3,label='3月份')
plt.scatter(x10,y10,label='10月份')
    #   plot 折现图，scatter散点图
    

#调整x轴刻度
_x = list(x3)+list(x10)
_x_labels = ['3月%i日'%i for i in x3]
_x_labels += ['10月%i日'%i for i in x10]

plt.xticks(_x[::3],_x_labels[::3],rotation=45)

plt.xlabel('时间')
plt.ylabel('温度')
plt.title('3月10月份温度散点对比')

plt.legend(loc='upper left')

plt.show()
```
